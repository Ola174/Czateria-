```
% Deklaracja dynamicznego predykatu
:- dynamic produkt/4.

% Fakty definiujące produkty w magazynie (z kategoriami)
produkt(tv, 10, sektor_a, elektronika).
produkt(pralka, 5, sektor_b, agd).
produkt(lodowka, 3, sektor_c, agd).
produkt(odtwarzacz, 7, sektor_b, elektronika).

% Reguła sprawdzająca, czy produkt jest dostępny w magazynie
dostepny_produkt(X) :-
    produkt(X, Ilosc, _, _),
    Ilosc > 0.

% Reguła sprawdzająca, czy produkt jest dostępny w danym sektorze
dostepny_w_sektorze(X, Sektor) :-
    produkt(X, Ilosc, Sektor, _),
    Ilosc > 0.

% Reguła przenosząca produkt do innego sektora
przenies_produkt(X, NowySektor) :-
    produkt(X, Ilosc, StarySektor, Kategoria),
    retract(produkt(X, Ilosc, StarySektor, Kategoria)),   % Usuwamy produkt ze starego sektora
    assertz(produkt(X, Ilosc, NowySektor, Kategoria)),    % Dodajemy produkt do nowego sektora
    format('Produkt ~w został przeniesiony z ~w do ~w.~n', [X, StarySektor, NowySektor]).

% Reguła uzupełniania stanu produktu
uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor, Kategoria),
    NowaIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor, Kategoria)),
    assertz(produkt(X, NowaIlosc, Sektor, Kategoria)),
    format('Ilość produktu ~w została zwiększona o ~w, nowy stan to ~w.~n', [X, N, NowaIlosc]).

% Reguła usuwania produktu z magazynu
usun_produkt(X) :-
    retractall(produkt(X, _, _, _)),
    format('Produkt ~w został usunięty z magazynu.~n', [X]).

% Reguła generowania raportu o stanie magazynu
raport_magazynu :-
    findall((X, Ilosc, Sektor, Kategoria), produkt(X, Ilosc, Sektor, Kategoria), Lista),
    print_raport(Lista).

% Pomocnicza reguła do formatowania raportu
print_raport([]).
print_raport([(X, Ilosc, Sektor, Kategoria) | T]) :-
    format('Produkt: ~w, Ilość: ~w, Sektor: ~w, Kategoria: ~w~n', [X, Ilosc, Sektor, Kategoria]),
    print_raport(T).


% Przykładowe zapytania:
% Zapytania o dostępny produkt
?- dostepny_produkt(tv). % Oczekiwania: true
?- dostepny_produkt(pralka). % Oczekiwania: true
?- dostepny_produkt(lodowka). % Oczekiwania: true
?- dostepny_produkt(telefon). % Oczekiwania: false

% Zapytania o przeniesiony produkt
?- przenies_produkt(tv, sektor_b). % Oczekiwania: produkt(tv, 10, sektor_b, elektronika)
Opis: tv przeniesiono do sektora sektor_b.
?- przenies_produkt(pralka, sektor_c). % Oczekiwania: produkt(pralka, 5, sektor_c, agd)
Opis: pralkę przeniesiono do sektora sektor_c.
?- przenies_produkt(lodowka, sektor_a). % Oczekiwania: produkt(lodowka, 3, sektor_a, agd)
Opis: lodówkę przeniesiono do sektora sektor_a.
?- przenies_produkt(odtwarzacz, sektor_c). % Oczekiwania: produkt(odtwarzacz, 7, sektor_c, elektronika)
Opis: odtwarzacz przeniesiono do sektora sektor_c.

% Zapytania o uzupełnienie stanu
?- uzupelnij_stan(tv, 5). % Oczekiwania: produkt(tv, 15, sektor_a, elektronika)
Opis: Ilość produktu tv została zwiększona o 5, wynikowy stan to 15.
?- uzupelnij_stan(pralka, 3). % Oczekiwania: produkt(pralka, 8, sektor_b, agd)
Opis: Ilość produktu pralka została zwiększona o 3, wynikowy stan to 8.
?- uzupelnij_stan(lodowka, 2).% Oczekiwania: produkt(lodowka, 5, sektor_c, agd)
Opis: Ilość produktu lodówka została zwiększona o 2, wynikowy stan to 5.
?- uzupelnij_stan(odtwarzacz, 4).% Oczekiwania: produkt(odtwarzacz, 11, sektor_b, elektronika)
Opis: Ilość produktu odtwarzacz została zwiększona o 4, wynikowy stan to 11.
```

```
% Fakty o miastach
miasto(warszawa).
miasto(krakow).
miasto(gdansk).
miasto(wroclaw).
miasto(lodz).
miasto(rzeszow).

% Fakty o drogach (połączenia drogowe między miastami)
droga(warszawa, krakow, 4).
droga(krakow, gdansk, 3).
droga(warszawa, gdansk, 7).
droga(warszawa, wroclaw, 5).
droga(wroclaw, krakow, 3).
droga(gdansk, wroclaw, 6).
droga(wroclaw, lodz, 2).
droga(lodz, rzeszow, 4).
droga(rzeszow, gdansk, 6).
droga(gdansk, lodz, 6).
droga(krakow, rzeszow, 5).
droga(warszawa, rzeszow, 6).
droga(lodz, krakow, 6).

% Reguła sprawdzająca, czy istnieje bezpośrednia droga (dwukierunkowa)

droga(X, Y) :-
    droga(X, Y, _).  % Sprawdzenie, czy istnieje bezpośrednia droga

droga(X, Y) :-
    droga(Y, X, _).  % Dodajemy symetrię, bo droga jest dwukierunkowa


% Reguła obliczająca czas przejazdu
czas_przejazdu(X, Y, T) :- bezposrednia_droga(X, Y, T).
czas_przejazdu(X, Y, T) :-
    bezposrednia_droga(X, Z, T1),
    czas_przejazdu(Z, Y, T2),
    T is T1 + T2,
    X \= Y.

% Reguła do wyznaczania najkrótszej trasy
najkrotsza_trasa(X, Y, NajkrotszaTrasa, NajkrotszyCzas) :-
    findall((Trasa, Czas), trasa(X, Y, Trasa, Czas), MozliweTrasy),
    MozliweTrasy \= [],
    minimalna_trasa(MozliweTrasy, (NajkrotszaTrasa, NajkrotszyCzas)).

% Reguła definiująca możliwe trasy
trasa(X, Y, [X, Y], Czas) :- droga(X, Y, Czas).
trasa(X, Y, [X | ResztaTrasy], Czas) :-
    droga(X, Z, Czas1),
    \+ member(Z, [X | ResztaTrasy]), % Unikamy powtórzeń w trasie
    trasa(Z, Y, ResztaTrasy, Czas2),
    Czas is Czas1 + Czas2.

% Reguła do znalezienia trasy o minimalnym koszcie
minimalna_trasa([Trasa], Trasa). % Przypadek bazowy: tylko jedna trasa
minimalna_trasa([(T1, C1), (T2, C2) | Reszta], Min) :-
    (C1 =< C2 ->
        minimalna_trasa([(T1, C1) | Reszta], Min)
    ;
        minimalna_trasa([(T2, C2) | Reszta], Min)
    ).


% Zapytania testowe
% 1. Sprawdzenie, czy istnieje bezpośrednia droga między Warszawą a Krakowem:
?- droga(warszawa, krakow). % Oczekiwany wynik: true

% 2. Określenie czasu przejazdu między Warszawą a Gdańskiem:
?- czas_przejazdu(warszawa, gdansk, T). % Oczekiwany wynik: T = 7

% 3. Wyznaczenie najkrótszej trasy z Warszawy do Łodzi:
?- najkrotsza_trasa(lodz, rzeszow, Trasa, Czas). % Oczekiwany wynik: Trasa = [lodz, rzeszow] Czas= 4
% 4. Wyznaczenie najkrótszej trasy z Gdańska do Rzeszowa:
?- najkrotsza_trasa(gdansk, rzeszow, Trasa, Czas). % Oczekiwany wynik: Trasa = [gdansk, rzeszow], Czas = 6

% 5. Określenie czasu przejazdu między Krakowem a Rzeszowem:
?- czas_przejazdu(krakow, rzeszow, T). % Oczekiwany wynik: T = 5

% 6. Wyznaczenie najkrótszej trasy z Krakowa do Gdańska:
?- najkrotsza_trasa(krakow, gdansk, Trasa, Czas). % Oczekiwany wynik: Trasa = [krakow, gdansk], Czas = 3
```
